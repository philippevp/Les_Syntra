# Python – *args


```
def myfunc(*args):
 for x in args:
 print(x)
```



*   *args gebruiken we om een ongekend aantal items door te geven aan een functie, *args is altijd een tuple
Je mag uiteraard args vervangen door je zelf gekozen naam, alleen de * is belangrijk om dit aan te geven
*   Waarom zou je dit gebruiken ? Om bijvoorbeeld een lijst van getallen door te geven, of een lijst van
punten voor een grafiek

````python
def myfunc(*args):
    print(type (args))
    for x in args:
        print(x)

myfunc(4, 5, "abc", [1, 3])
````
